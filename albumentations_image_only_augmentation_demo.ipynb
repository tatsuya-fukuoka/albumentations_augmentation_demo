# 画像のデータ拡張（bbox反映無しver.）

```python
# ライブラリのインストール
import albumentations as A
import cv2

# データ拡張パイプラインの定義
transform = A.Compose([
    A.RandomCrop(width=416, height=416),
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.2),
])
```

```python
import cv2
image = cv2.imread("/content/dog.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
import cv2
from PIL import Image
from IPython.display import display

pil_img = Image.fromarray(image)
display(pil_img)
```

```python
transformed = transform(image=image)
transformed_image = transformed["image"]
pil_trans_img = Image.fromarray(transformed_image)
display(pil_trans_img)
```

```python
transformed2 = transform(image=image)
transformed_image2 = transformed2["image"]
pil_trans_img2 = Image.fromarray(transformed_image2)
display(pil_trans_img2)
```
